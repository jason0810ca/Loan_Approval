```python
df_categorical = df.loc[:,['Securities Account', 'CD Account', 'Online', 'CreditCard']]
df_numerical = df.loc[:,['Age', 'Experience', 'Income', 'CCAvg', 'Mortgage']]
df_ordinal = df.loc[:,['Family', 'Education']]
y = df.loc[:,'Personal Loan']
```

```python
df_categorical_dummy = pd.get_dummies(df_categorical, drop_first=True)
df_categorical_dummy = df_categorical_dummy.astype(int)
```

```python
df_categorical_dummy.head()
```

```python
df_numerical.head()
```

```python
sc = StandardScaler()
```

```python
df_numerical_sc = sc.fit_transform(df_numerical)
```

```python
df_numerical_sc = pd.DataFrame(df_numerical_sc, columns=df_numerical.columns)
```

```python
df_numerical_sc.head()
```

```python
for column in df_ordinal.columns:
  x = df_ordinal.loc[:,column].unique()
  print(column, x)
```

```python
family_mapping = pd.Series()
i = 0
for category in np.unique(df.loc[:,'Family']):
  family_mapping.at[category] = i
  i += 1

family_mapping = family_mapping.to_dict()
family_mapping

education_mapping = pd.Series()
i = 0
for category in np.unique(df.loc[:,'Education']):
  education_mapping.at[category] = i
  i += 1

education_mapping = education_mapping.to_dict()
education_mapping
```

```python
type(family_mapping)
```

```python
df_ordinal.loc[:,'Family'] = df_ordinal.loc[:,'Family'].map(family_mapping)
df_ordinal.loc[:,'Education'] = df_ordinal.loc[:,'Education'].map(education_mapping)

df_ordinal.head()
```

```python
X_ready_to_model = pd.concat([df_categorical_dummy, df_numerical_sc, df_ordinal], axis=1)
```

```python
X_ready_to_model.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ready_to_model, y, test_size=0.2, random_state=42)

X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
train_df = pd.concat([X_train, y_train], axis=1)
```

```python
plt.barh(np.unique(train_df.loc[:,'Personal Loan']), train_df.loc[:,'Personal Loan'].value_counts())
plt.yticks(np.unique(train_df.loc[:,'Personal Loan']), fontsize=15)
plt.xticks(fontsize=15)
for index, value in train_df.loc[:,'Personal Loan'].value_counts().sort_index().items():
  plt.text(value, index, f'{value}', color='red', fontsize=15)
plt.xlabel('Counts', fontsize=15)
plt.ylabel('Labels', fontsize=15)
plt.title('Counts for the Two Labels', fontsize=15, loc='left')
ax = plt.gca()
ax.set_frame_on(False)
```

```python
train_df_1 = train_df[train_df.loc[:,'Personal Loan'] == 1].reset_index(drop=True)
train_df_0 = train_df[train_df.loc[:,'Personal Loan'] == 0].reset_index(drop=True)
```

```python
train_df_1
```

```python
train_df_balanced = pd.concat([train_df_1]*10 + [train_df_0], axis=0)
```

```python
X_train_df_balanced = train_df_balanced.drop('Personal Loan', axis=1)
y_train_df_balanced = train_df_balanced.loc[:,'Personal Loan']
```

```python
train_df_balanced.loc[:,'Personal Loan'].value_counts()
```

```python
df_numerical_sc_feature_selection = df_numerical_sc.copy()

def calc_vif(X):

    vif = pd.DataFrame()
    vif["Features"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

vif_result = calc_vif(df_numerical_sc_feature_selection)
vif_result
```

```python
vif_result.dtypes
```

```python
for feature in range(len(df_numerical_sc_feature_selection.columns)):
  if np.max(vif_result.loc[:,'VIF'])>10:
    removed_index = np.argmax(vif_result.loc[:,'VIF'])
    removed_feature = vif_result.iloc[removed_index, 0]
    df_numerical_sc_feature_selection = df_numerical_sc_feature_selection.drop(removed_feature, axis=1)
    vif_result = calc_vif(df_numerical_sc_feature_selection)

vif_result
```

```python
df_feature_selection = pd.concat([df_numerical_sc_feature_selection, df_categorical_dummy, df_ordinal], axis=1)
```

```python
features_to_remove = []

for feature in X_train_df_balanced.columns:
  if feature in df_feature_selection:
    features_to_remove.append(True)
  else:
    features_to_remove.append(False)

features_to_remove
```

```python
X_train_df_balanced = X_train_df_balanced[X_train_df_balanced.columns[features_to_remove]]
X_test = X_test[X_test.columns[features_to_remove]]
```

```python
X_test
```

```python
model = LogisticRegression(solver= "liblinear", random_state = 42)
```

```python
sfs = SFS(model,
      k_features="best",  # Select the best number of features
      forward=True,        # Forward selection
      floating=False,      # Not Floating steps
      cv=5,                # 5-fold cross-validation
      verbose=1)           # Show progress

# Fit the SFS model on the training data
sfs = sfs.fit(X_train_df_balanced, y_train_df_balanced)

# Get the names of the selected features
selected_features = list(sfs.k_feature_names_)
print("Selected features:", selected_features)
```

```python
X_train_df_balanced = X_train_df_balanced[selected_features]
X_test = X_test[selected_features]
```

```python
model.fit(X_train_df_balanced, y_train_df_balanced)
```

```python
y_pred = model.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.xlabel('Predicted', fontsize=15)
plt.ylabel('True', fontsize=15)
plt.title('Confusion Matrix', loc='left')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
